```python
#01.How do you create a 2D NumPy array and calculate the sum of each row?

import numpy as np

# Create a 2D NumPy array
my_array = np.array([[1, 2, 3], [4, 5, 6]])

# Calculate the sum of each row
row_sums = np.sum(my_array, axis=1)

# Print the row sums
print(row_sums)

```

```python
#02.Write a Pandas script to find the mean of a specific column in a DataFrame?

import pandas as pd
import numpy as np

# Sample DataFrame (replace with your actual data)
data = {'col_A': [1, 2, 3, 4, 5],
        'col_B': [6, 7, 8, np.nan, 10],
        'col_C': [11, 12, 13, 14, 15]}
df = pd.DataFrame(data)

# Calculate the mean of 'col_B', handling potential NaN values
mean_col_b = df['col_B'].mean(skipna=True)

# Print the result
print(f"The mean of 'col_B' is: {mean_col_b}")
```

```python
#04.How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your data is in a DataFrame called 'df'
# Replace this with your actual data loading code
data = pd.DataFrame({'A': [1, 2, 3, 4, 5],
                     'B': [5, 4, 3, 2, 1],
                     'C': [2, 4, 6, 8, 10]})
corr = data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#07.Write a program to perform element-wise multiplication of two NumPy arrays?


import numpy as np

def multiply_arrays(array1, array2):
    """
    Performs element-wise multiplication of two NumPy arrays.
     Args:
        array1 (numpy.ndarray): The first array.
        array2 (numpy.ndarray): The second array.
     Returns:
        numpy.ndarray: A new array containing the element-wise products, or None if the arrays
                       cannot be multiplied due to shape mismatch.
    """
    if array1.shape != array2.shape:
        print("Arrays must have the same shape for element-wise multiplication.")
        return None
    return np.multiply(array1, array2)

if __name__ == "__main__":
    array1 = np.array([[1, 2, 3], [4, 5, 6]])
    array2 = np.array([[7, 8, 9], [10, 11, 12]])

    result_array = multiply_arrays(array1, array2)

    if result_array is not None:
        print("Array 1:")
        print(array1)
        print("Array 2:")
        print(array2)
        print("Element-wise multiplication result:")
        print(result_array)

    array3 = np.array([1, 2, 3])
    array4 = np.array([4, 5])
    multiply_arrays(array3, array4)
```

```python
#08.Create a line plot with multiple lines using Matplotlib?

import matplotlib.pyplot as plt

# Sample data for multiple lines
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 1, 3, 5]
y2 = [3, 1, 4, 2, 5]
y3 = [1, 3, 5, 4, 2]

# Plotting the lines
plt.plot(x, y1, label='Line 1', marker='o')
plt.plot(x, y2, label='Line 2', marker='x')
plt.plot(x, y3, label='Line 3', marker='s')

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Multiple Lines')

# Adding a legend
plt.legend()

# Adding grid lines
plt.grid(True)

# Displaying the plot
plt.show()
```

```python
#09.Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold?

import pandas as pd
import numpy as np

# Create a sample DataFrame
data = {'col1': np.random.randint(1, 100, 10),
        'col2': np.random.rand(10),
        'col3': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C', 'A']}
df = pd.DataFrame(data)

# Define the threshold
threshold = 50

# Filter rows where 'col1' is greater than the threshold
filtered_df = df[df['col1'] > threshold]

# Print the original and filtered DataFrames
print("Original DataFrame:")
print(df)
print("\nFiltered DataFrame (col1 > {}):".format(threshold))
print(filtered_df)
```

```python
#13.Create a 3D scatter plot using Plotly.

import plotly.graph_objects as go
import numpy as np

# Generate random data for the scatter plot
np.random.seed(42)
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)
colors = np.random.rand(100)
sizes = np.random.randint(10, 20, 100)

# Create the 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=sizes,
        color=colors,
        opacity=0.8,
        colorscale='Viridis'
    )
)])

# Update layout for better visualization (optional)
fig.update_layout(
    margin=dict(l=0, r=0, b=0, t=0),
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    )
)

# Display the plot
fig.show()
```
